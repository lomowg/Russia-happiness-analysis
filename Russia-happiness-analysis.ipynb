# Загружаем датасеты <a class="anchor" id="1-bullet"></a>

```python
df_2015 = pd.read_csv('data-sets/2015.csv')
df_2016 = pd.read_csv('data-sets/2016.csv')
df_2017 = pd.read_csv('data-sets/2017.csv')
df_2018 = pd.read_csv('data-sets/2018.csv')
df_2019 = pd.read_csv('data-sets/2019.csv')
df_2015.head(10)
```

```python
df = pd.read_csv('data-sets/world_bank_development_indicators.csv')
df.head(10)
```

# Оставляем нужные данные

Найдём данные только по России

```python
df_russia = df[df['country'] == 'Russian Federation'].rename(columns={'date': 'year'})
df_russia
```

Преобразуем данные столбца year в тип данных datetime и оставим только год

```python
df_russia['year'] = pd.to_datetime(df_russia['year']).dt.year
df_russia
```

Оставим итересующие нас года (2015-2019)

```python
df_russia = df_russia[(2015 <= df_russia['year']) & (df_russia['year'] <= 2019)]
df_russia
```

# Избавляемся от пропусков

Мы видим, что у нас есть пропуски и их много

```python
cols_with_nans = df_russia.isna().sum()
cols_with_nans = cols_with_nans[cols_with_nans != 0]
cols_with_nans
```

В данной ситуации можно дропнуть столбцы, где у нас в итересующих нас годах есть пропуски, но сначала я попытаюсь их заполнить средними значениями по столбцам всего датасета. Для этого проверим нормальность распределения этих значений, иначе заполнение средним - очень плохая идея

```python
for col in cols_with_nans.index:
    sms.qqplot(df[col], line='s')
    plt.title(col)
    plt.show()
```

Как мы видим, все данные распределены далеко не нормально, поэтому лучше мы просто дропнем эти столбцы из нашего датафрейма

```python
df_russia = df_russia.dropna(axis=1)
df_russia.isna().sum()
```

# Загрузми второй датасет и объединим с первым

```python
russia_2015 = df_2015[df_2015['Country'] == 'Russia'][['Happiness Rank', 'Happiness Score']]
russia_2016 = df_2016[df_2016['Country'] == 'Russia'][['Happiness Rank', 'Happiness Score']]
russia_2017 = df_2017[df_2017['Country'] == 'Russia'][['Happiness.Rank', 'Happiness.Score']]
russia_2018 = df_2018[df_2018['Country or region'] == 'Russia'][['Overall rank', 'Score']]
russia_2019 = df_2019[df_2019['Country or region'] == 'Russia'][['Overall rank', 'Score']]

df_rank = pd.concat([russia_2015['Happiness Rank'], russia_2016['Happiness Rank'], russia_2017['Happiness.Rank'], russia_2018['Overall rank'], russia_2019['Overall rank']]).to_frame()\
                                                                                     .rename(columns={0: 'rank'})\
                                                                                     .reset_index(drop=True)
df_score = pd.concat([russia_2015['Happiness Score'], russia_2016['Happiness Score'], russia_2017['Happiness.Score'], russia_2018['Score'], russia_2019['Score']]).to_frame()\
                                                                                     .rename(columns={0: 'score'})\
                                                                                     .reset_index(drop=True)

df_rank['year'] = [2015, 2016, 2017, 2018, 2019]
df_score['year'] = [2015, 2016, 2017, 2018, 2019]
df_merged = df_rank.merge(df_score)
df_russia = df_russia.merge(df_merged)
df_russia
```

```python
correlation_matrix = df_russia.corr(numeric_only=True)
variables_to_drop = correlation_matrix[abs(correlation_matrix['score']) < 0.5].index
df = df_russia.drop(variables_to_drop, axis=1)
df
```
